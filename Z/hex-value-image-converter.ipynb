```python
# Read the hex data from the file
with open('image_hex.txt', 'r') as file:
    hex_data = file.read()

print(hex_data)
# Decode the hex data into binary
binary_data = bytes.fromhex(hex_data)

# Write the binary data to a new file with .jpg extension
with open('output.jpg', 'wb') as file:
    file.write(binary_data)
```
